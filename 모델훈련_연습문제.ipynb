## **| 모델 훈련 연습 문제**
___
- 출처 : 핸즈온 머신러닝 Ch04 연습문제 1, 5, 9, 10번
- 개념 문제의 경우 텍스트 셀을 추가하여 정답을 적어주세요.

### **1. 수백만 개의 특성을 가진 훈련 세트에서는 어떤 선형 회귀 알고리즘을 사용할 수 있을까요?**
___

확률적 경사 하강법(SGD), 미니배치 경사 하강법

배치 경사 하강법은 훈련 세트가 메모리 크기에 맞을 때 사용할 수 있다.

### **2. 배치 경사 하강법을 사용하고 에포크마다 검증 오차를 그래프로 나타내봤습니다. 검증 오차가 일정하게 상승되고 있다면 어떤 일이 일어나고 있는 걸까요? 이 문제를 어떻게 해결할 수 있나요?**
___

모델이 훈련 세트에 과대적합되었다. 따라서, 학습률을 낮추거나 조기종료 파라미터를 조정한다.

### **3. 릿지 회귀를 사용했을 때 훈련 오차가 검증 오차가 거의 비슷하고 둘 다 높았습니다. 이 모델에는 높은 편향이 문제인가요, 아니면 높은 분산이 문제인가요? 규제 하이퍼파라미터 $\alpha$를 증가시켜야 할까요 아니면 줄여야 할까요?**
___

이 모델은 훈련 세트에 과소적합된 것으로, 높은 편향이 문제이다. 규제 하이퍼파라미터 $\alpha$를 감소시켜야 한다.

### **4. 다음과 같이 사용해야 하는 이유는?**
___
- 평범한 선형 회귀(즉, 아무런 규제가 없는 모델) 대신 릿지 회귀
- 릿지 회귀 대신 라쏘 회귀
- 라쏘 회귀 대신 엘라스틱넷

- 규제가 있는 모델이 아무런 규제가 없는 모델보다 일반적으로 성능이 더 좋기 때문에 릿지 회귀를 선호한다.

- 라쏘 회귀는 l1페널티를 이용하여 덜 중요한 특성의 가중치를 제거하도록 자동으로 특성을 선택하고 희소 모델을 만든다. 따라서, 몇 개의 특성만 실제 유용할 것이라고 의심이 될 때 라쏘 회귀를 사용한다. 하지만, 확신이 없다면 릿지 회귀를 사용해야 한다.

- 엘라스틱넷은 릿지 회귀와 라쏘 회귀를 절충한 모델로, 불필요한 특성의 가중치를 0으로 만들어 특성수가 훈련 샘플 수보다 많거나 특성 몇 개가 강하게 연관되어 있을 때는 라쏘 회귀가 불규칙하게 행동하므로 엘라스틱넷을 선호한다. 불규칙한 행동이 없는 라쏘 회귀를 원한다면 엘라스틱넷에 l1_ratio를 1에 가깝게 설정하면 된다.
